# 중복 아이디 제거, 질병정도 max값으로 통합

```python
user_df['id_new'] = user_df['id'].apply(lambda x: x[:-2])
user_df['id_new']
```

```python
user_df[user_df['id_new']=='0013_A2LEBJJDE00060O_1602578303771']
```

```python
user_df.columns
# '미세각질', '피지과다', '모낭사이홍반', '모낭홍반농포', '비듬','탈모'
```

```python
disease = ['미세각질', '피지과다', '모낭사이홍반', '모낭홍반농포', '비듬','탈모']
```

```python
user_df[user_df['id_new']=='0013_A2LEBJJDE00060O_1602578303771'][['미세각질', '피지과다', '모낭사이홍반', '모낭홍반농포', '비듬','탈모']]
```

```python
user_df[user_df['id_new']=='0013_A2LEBJJDE00060O_1602578303771']['미세각질'].max()
```

```python
user_df[user_df['id_new']=='0013_A2LEBJJDE00060O_1602578303771'][['미세각질', '피지과다', '모낭사이홍반', '모낭홍반농포', '비듬','탈모']] = [0,2,3,0,1,0]
```

```python
user_df['id_new'].unique()
```

```python
# 아이디값을 중복없이 만들어 리스트에 저장
id_list = list(user_df['id_new'].unique())
id_list
```

```python
# 하나의 아이디에 대하여 질병 수치 최대값만 존재하는 리스트 만들기
duplicated_list =[]
for ids in id_list:
    dis_value =[]
    for dis in disease:
        dis_value.append(user_df[user_df['id_new']==ids][dis].max())
    duplicated_list.append(dis_value)
duplicated_list
```

```python
# 리스트 데이터프레임화
duplicated_df = pd.DataFrame(duplicated_list, columns = disease)
duplicated_df
```

```python
# 기존의 id가 중복된 값들을 첫번째 값 제외 모두 제거
user_df_01 = user_df.drop_duplicates(['id_new'], keep = 'first')
user_df_01.info()
```

```python
# 기존 id, 질병 값 제거
user_df_01.drop(['id','id_new'], axis=1,inplace=True)
user_df_01.info()
```

```python
user_df_01[disease].head()
```

```python
user_df_01.drop(disease,axis = 1, inplace = True)
user_df_01.head()
```

```python
duplicated_df.head()
```

```python
# 새 질병값 추가하기 위헤 index 리셋
# index 리셋하지 않으면 리셋되지 않은 index에 값이 들어감
user_df_01.reset_index(inplace=True)
user_df_01.head()
```

```python
# 한 사람에 대해 max 시킨 질병값 columns 추가
user_df_01[disease] = duplicated_df
user_df_01.head()
```

```python
user_df_01.info()
```

```python
# 불필요한 column 제거
user_df_01.drop(['index'],axis=1,inplace=True)
user_df_01.head()
```

```python
user_df = user_df_01
user_df.head()
```

# 질병 columns 추가

```python
user_df['건성'] = [1 if m>1 else 0 for m in user_df['미세각질']]
user_df['지성'] = [1 if m>1 else 0 for m in user_df['피지과다']]
user_df['민감성'] = [1 if m>1 else 0 for m in user_df['모낭사이홍반']]
user_df['염증성'] = [1 if m>1 else 0 for m in user_df['모낭홍반농포']]
user_df['비듬성'] = [1 if m>1 else 0 for m in user_df['비듬']]
user_df['탈모성'] = [1 if m>1 else 0 for m in user_df['탈모']]
```
