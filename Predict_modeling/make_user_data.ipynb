```python
file_path="./data/Meta/"
```

```python
data_low = []
for file in json_list:
    with open(file_path+file,'r', encoding='utf-8') as j:
        contents = json.loads(j.read())  # open : r - 읽기모드, w-쓰기모드, a-추가모드 
        low = list(contents.values())
        del low[2]
        del low[2:len(low):2]
        low.insert(0,file[:-13])
        data_low.append(low)
        
data_low
```

```python
Meta_df_kr = pd.DataFrame(data_low, columns = columns_kr)
Meta_df_kr
```

```python
label_columns = ['id','미세각질','피지과다','모낭사이홍반','모낭홍반농포','비듬','탈모']
label_columns
```

```python
label_file_path = './data/Label/'
```

```python
label_json_list= os.listdir('./data/Label')
len(label_json_list)
```

```python
label_data = []
for file in label_json_list:
    with open(label_file_path+file,'r', encoding='utf-8') as j:
        contents = json.loads(j.read())  # open : r - 읽기모드, w-쓰기모드, a-추가모드 
        low = list(contents.values())
        del low[1]
        label_data.append(low)
        
label_data
```

```python
Label_df = pd.DataFrame(label_data, columns = label_columns)
Label_df
```

```python
Label_df.to_csv('./data/label_kr.csv', index=False)
```
