```python
# 모듈 및 라이브러리 호출
import os
from tqdm import tqdm
from tensorflow.keras.preprocessing.image import ImageDataGenerator,img_to_array, load_img
```

```python
# 해당 경로 폴더의 이미지 갯수 확인
# 확인 후 가장 큰 값을 반환
img_count = [len(os.listdir('./'+path)) for path in os.listdir('./')]
img_count.sort(reverse=True)
count_max = img_count[0]
print(img_count,'\n',count_max)
# 
```

```python
def aug_img(path,aug_num):# 해당 경로 내 모든 폴더의 이미지를 최소 aug_num만큼 생성
# 이미지 제너레이터 객체 생성
    gen = ImageDataGenerator(
    zoom_range = [0.8,1],# 0.8~1 범위 내에서 확대
    horizontal_flip = True,# 50% 확률로 수평 반전
    vertical_flip = True,# 50% 확률로 수직 반전
    brightness_range = [0.7,1.3] # 70% ~ 130% 범위로 밝기 조절
    )

    folder_list = [path + file for file in os.listdir(path) if not file.endswith('.ini')] # 폴더만 추출

    for folder in tqdm((folder_list)): # 모든 폴더에 대해 반복
        iter_num = (aug_num//(len(os.listdir(folder)))) # 이미지당 생성 횟수 지정
        if iter_num > 1: # 몫이 1 이하면 aug 시행하지 않음
            for img in os.listdir(folder):# 모든 이미지에 대해 반복
                raw_img = load_img(os.path.join(folder+'/'+img))# 이미지 불러오기
                x = img_to_array(raw_img)# 이미지를 array로 변경
                x = x.reshape((1,)+x.shape)# img를 4차원으로
                i = 0# 생성된 이미지 카운트를 위한 변수
                # 동일 폴더 내에 이미지 생성하기
                for batch in gen.flow(x,batch_size = 1, save_to_dir=folder, save_prefix="gen_1", save_format="jpg"):
                    i += 1
                    if i>= iter_num:# 생성횟수만큼 생성하면 종료
                        break
```

```python
# 함수 실행
aug_img('./',count_max)
```
